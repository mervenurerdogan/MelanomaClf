```python
path=Path("/content/train")
path.ls()
```

```python
train_fnames=get_image_files(path)
```

```python



dblock=DataBlock(blocks=(ImageBlock(cls=PILImage),CategoryBlock),
                 splitter=RandomSplitter(valid_pct=0.2),
                 get_y=parent_label,
                 item_tfms=Resize(224,method='squish'),
                 batch_tfms=[*aug_transforms(size=224, max_warp=0,do_flip=True,
                      flip_vert=True),Normalize.from_stats(*imagenet_stats)]
                 
)
```

```python
dls=dblock.dataloaders(train_fnames)
```

```python
dls.c,len(dls.train_ds),len(dls.valid_ds)
```

```python
dls.vocab
```

```python
dls.show_batch(max_n=12, nrows=4,figsize=(15,15))
```

```python
learn = vision_learner(dls, 'convnext_tiny_in22k', metrics=accuracy)
```

```python
print(learn.summary())
```

```python
learn.lr_find()
```

```python
learn.fine_tune(40,0.0005)
```

```python
learn.recorder.plot_loss()
```

```python
interp=ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.print_classification_report()
```

```python
preds, targs = learn.tta()
accuracy(preds, targs).item()
```

```python
interp.most_confused()
```

```python
interp.plot_top_losses(1, figsize=(10,5))
```

```python
learn.show_results()
```

```python
learn.save(path/'Colon_ConvNeXt_Train')
```
